```python
import pandas as pd  # import library 
import matplotlib.pyplot as plt
```

```python
import warnings as wr  # import warnings module for ignore warning
wr.filterwarnings("ignore")
```

```python
df = pd.read_csv("Height-Weight Data.csv") # read the csv file
```

```python
df.head() # it shows 5 rows
```

```python
df.shape # it shows how many row and column we have.
```

```python
df.isnull().sum() # it shows how many missing values we have.
```

```python
df.info() # as like as isnull().sum() but it shows in details.
```

```python
plt.scatter(df.Height,df.Weight,color="Green")# it shows all ponts of dataset
plt.xlabel("Height in C.M") # anything you can write, it willbe showing for X-axis
plt.ylabel("weight in POUND") # anything you can write , it willbe showing for Y-axis
plt.title("Height-weight plot") # for caption or title
```

```python
df.Height.mean() # X-mean
```

```python
df.Weight.mean() # Y-mean
```

```python
x = df[["Height"]]
```

```python
y= df[["Weight"]]
```

```python
x.head() # first 5  height value 
```

```python
y.head() # first 5 weight value
```

```python
from sklearn.linear_model import LinearRegression # just import Linear Regression Module from sklearn library
```

```python
regg=LinearRegression() # creating object for Linear Regression
```

```python
regg.fit(x,y) #training the model
```

```python
regg.coef_ # the value of m in the equation of y=mx+c
```

```python
regg.intercept_ # the value of c in the equation of y=mx+c
```

```python
regg.predict([[70]]) # it shows predicted weight, when we give height 
```

```python
# y = mx+c
result=((regg.coef_)*70) + regg.intercept_
result # manually process
```

```python
regg.predict(x) # predict all weight from dataset
```

```python
df["predicted_Weight"]= regg.predict(x) #add a new column name "predict_Weight"
```

```python
df.head()
```

```python
plt.plot(df.Height.squeeze(), regg.predict(df[["Height"]])) # it shows the predict main result's line
plt.scatter(df.Height,df.Weight,color="Green")# it shows all ponts of dataset
plt.xlabel("Height in C.M") # anything you can write, it willbe showing for X-axis
plt.ylabel("weight in POUND") # anything you can write , it willbe showing for Y-axis
plt.title("Height-weight plot") # for caption or title
```

```python
regg.score(x,y) # training accuracy
```

```python
from sklearn.metrics import mean_squared_error # import MSE from sklearn
```

```python
mean_squared_error(df[["Height"]], regg.predict(df[["Height"]])) #MSE (Mean Squre Error) value
```

```python
mean_squared_error(x,df.predicted_Weight) # df[["Height"]] = x
```
