```python
x=df[["area"]] # X ke alltime 2D rakte hobe. tai 2 ta squre bracket use korte hobe.
```

```python
# x = df.drop("rent" , axis=1) (eita direct way..uporer ta manually process)
```

```python
y=df[["rent"]]
```

```python
# y = df.drop("area", axis=1) (eita direct way ..uporer ta manually process)
```

```python
y.head()
```

```python
x.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression() # creating object for liner regression
```

```python
reg.fit(x,y) #training the model
```

```python
reg.coef_ # y=mx+c equation e m er value
```

```python
reg.intercept_ #y=mx+c equation e c er value
```

```python
import warnings as wr # warnings module for ignore the warning
wr.filterwarnings("ignore")
```

```python
reg.predict([[4000]])
```

```python
reg.predict([[3000]])
```

```python
reg.predict([[8000]])
```

```python
# y = mx + c
result= ((reg.coef_) * 8000) + reg.intercept_
result # manually
```

```python
reg.predict(x)
```

```python
df["predicted_rent"] = reg.predict(x)
```

```python
df.head()
```

```python
plt.plot(df.area,reg.predict(x))
```

```python
plt.plot(df.area, reg.predict(df[["area"]]))
# X axis = area
# y axis = predict rent

plt.scatter(df.area, df.rent, color="green") 
plt.xlabel("Area in square feet")
plt.ylabel("Rent in Bangladesh Taka") 
plt.title("Area -house Rent")
```

```python
plt.plot(df.area.squeeze(), reg.predict(df[["area"]])) # jodi pc te version or dimension error ase tokhon squeeze() dilei hobe.
# X axis = area
# y axis = predict rent

plt.scatter(df.area, df.rent, color="green") 
plt.xlabel("Area in square feet")
plt.ylabel("Rent in Bangladesh Taka") 
plt.title("Area -house Rent")
```

```python
reg.score(x,y) # training accuracy
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(df[["area"]], reg.predict(df[["area"]]))
```

```python
mean_squared_error(x,df.predicted_rent) #Mean Squared Error(MSE)
```
